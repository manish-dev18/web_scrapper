```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://quotes.toscrape.com/"
```

```python
response = requests.get(url)
```

```python
html_text = bs(response.text, 'html.parser')
```

```python
bigbox = html_text.find_all("div", {"class": "col-md-8"})
```

```python
bigbox
```

```python
quotes = bigbox[1].find_all("span", {"class" : "text"})
```

```python
for i in quotes:
    print(i.text)
```

```python
authors = bigbox[1].find_all("small", {"class":"author"})
```

```python
for i in authors:
    print(i.text)
```

```python
tags = bigbox[1].find_all("a", {"class":"tag"})
```

```python
for i in tags:
    print(i.text)
```
